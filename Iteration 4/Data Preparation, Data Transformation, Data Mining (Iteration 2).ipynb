```python
import numpy 
import pandas as pd
```

```python
averageHourlyEarnings = pd.read_csv ('Datasets/Average Hourly Earnings by Industry ANZSIC06 and Sex Quarterly December 2018.csv')
averageWeeklyPaidHours = pd.read_csv ('Datasets/Average Weekly Paid Hours (Employees) by Industry (ANZSIC06) and Sex (Qrtly-MarJunSepDec).csv')
filledJobs = pd.read_csv ('Datasets/Filled Jobs by Industry (ANZSIC06) by Sex and Status in Employment (Qrtly-MarJunSepDec).csv')
fullTimeEquivalents = pd.read_csv ('Datasets/Full-Time Equivalent Employees by Industry (ANZSIC06) and Sex (Qrtly-MarJunSepDec).csv')
totalGrossEarnings = pd.read_csv ('Datasets/Total Gross Earnings by Industry (ANZSIC06) (Qrtly-MarJunSepDec).csv')
```

```python
averageHourlyEarnings_0 = averageHourlyEarnings[averageHourlyEarnings.Gender != 'Both Sexes']
averageWeeklyPaidHours_0 = averageWeeklyPaidHours[averageWeeklyPaidHours.Gender != 'Both Sexes']
filledJobs_0 = filledJobs[filledJobs.Gender != 'Both Sexes']
fullTimeEquivalents_0 = fullTimeEquivalents[fullTimeEquivalents.Gender != 'Both Sexes']
```

```python
averageHourlyEarnings_1 = averageHourlyEarnings_0.pivot_table(index=['YearQuarter', 'Industry'], columns='Gender')
averageWeeklyPaidHours_1 = averageWeeklyPaidHours_0.pivot_table(index=['YearQuarter', 'Industry'], columns='Gender')
filledJobs_1 = filledJobs_0.pivot_table(index=['YearQuarter', 'Industry'], columns='Gender')
fullTimeEquivalents_1 = fullTimeEquivalents_0.pivot_table(index=['YearQuarter', 'Industry'], columns='Gender')
```

```python
averageHourlyEarnings_2 = averageHourlyEarnings_1.assign(Average_Hourly_Pay_Difference = averageHourlyEarnings_1['Total (Ordinary Time + Overtime) Hourly Earnings', 'Male'] - averageHourlyEarnings_1['Total (Ordinary Time + Overtime) Hourly Earnings', 'Female'])
```

```python
averageHourlyEarnings_2
```

```python
averageHourlyEarnings_3 = averageHourlyEarnings_2.sort_values(['Industry', 'YearQuarter'])
averageHourlyEarnings_3
```

```python
averageHourlyEarnings_4 = averageHourlyEarnings_3.groupby(['Industry'])['Average_Hourly_Pay_Difference'].shift(-20)                                                                                                                            
averageHourlyEarnings_5 = pd.merge(averageHourlyEarnings_4, averageHourlyEarnings_3, on=['YearQuarter', 'Industry'])
averageHourlyEarnings_5 = averageHourlyEarnings_5.iloc[:, :-1]
averageHourlyEarnings_5
```

```python
a = pd.merge(averageHourlyEarnings_5, averageWeeklyPaidHours_1, on=['YearQuarter', 'Industry'])
b = pd.merge(a, filledJobs_1, on=['YearQuarter', 'Industry'])
c = pd.merge(b, fullTimeEquivalents_1, on=['YearQuarter', 'Industry'])
d = pd.merge(c, totalGrossEarnings, on=['YearQuarter', 'Industry'])
d
```

```python
d.index = pd.to_datetime(d.YearQuarter)
d = d.drop(['YearQuarter'], axis=1)
d1 = d.dropna(how='any',axis=0)
```

```python
X = d1.iloc[:,2:25]
X = X.astype(int)
Y = d1.iloc[:,1]
Y = Y.astype(int)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
model = ExtraTreesClassifier()
model.fit(X,Y)
print(model.feature_importances_) 
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
X
```

```python
# Load libraries
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
seed = 7
scoring = 'accuracy'
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
# evaluate each model in turn
results = []
names = []
for name, model in models:
	kfold = model_selection.KFold(n_splits=10, random_state=seed)
	cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
```

```python
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
```
