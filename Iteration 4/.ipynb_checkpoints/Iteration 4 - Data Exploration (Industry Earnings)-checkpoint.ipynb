```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('dataExploration').getOrCreate()
```

```python
data = spark.read.csv("Datasets/Total Gross Earnings by Industry (ANZSIC06) (Qrtly-MarJunSepDec).csv",inferSchema=True,header=True)
```

```python
data.show()
```

```python
results = data.filter("Industry = 'All Industries'").select('YearQuarter','Total Gross Earnings')
results.show()
```

```python
resultsX = data.filter("Industry = 'Electricity, Gas, Water and Waste Services'").select('YearQuarter','Total Gross Earnings')
resultsX.show()
```

```python
resultsY = data.filter("Industry = 'Manufacturing'").select('YearQuarter','Total Gross Earnings')
resultsY.show()
```

```python
results1 = results.toPandas()
```

```python
results3 = resultsX.toPandas()
```

```python
results5 = resultsY.toPandas()
```

```python
import matplotlib.pyplot as plt
```

```python
names = results1['YearQuarter']
values1 = results1['Total Gross Earnings']

fig, ax = plt.subplots()
ax.plot(names, values1)
ax.legend()

ax.set_xlabel('YearQuarter')
ax.set_ylabel('Total Gross Earnings ($M)')
ax.set_title('All Industries')

plt.xticks(rotation=90)

plt.rcParams["figure.figsize"] = (25,15)

plt.show()
```

```python
import matplotlib.pyplot as plt2
```

```python
names2 = results3['YearQuarter']
values3 = results3['Total Gross Earnings']

fig1, ax1 = plt2.subplots()
ax1.plot(names2, values3)
ax1.legend()

ax1.set_xlabel('YearQuarter')
ax1.set_ylabel('Total Gross Earnings ($M)')
ax1.set_title('Electricity, Gas, Water and Waste Services')

plt2.xticks(rotation=90)

plt2.rcParams["figure.figsize"] = (25,15)

plt2.show()
```
